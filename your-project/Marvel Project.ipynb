```python
list_of_columns = list(Marvel_FD.columns)
list_of_columns = list_of_columns[3:]
```

```python
def mean_df(df):
    dict_means = {}
    for i in list_of_columns:
        dict_means[i] = df[i].mean()
    return dict_means

mean_df(Marvel_DF)
```

```python
def median_df(df):
    dict_medians = {}
    for i in list_of_columns:
        dict_medians[i] = df[i].median()
    return dict_medians

median_df(Marvel_DF)
```
